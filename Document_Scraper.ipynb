```python
# Imports. If any fail to load, you may need to re/install them
import os
import requests
from urllib.parse import urljoin
from bs4 import BeautifulSoup
```

### Downloading PDFs from a Website

```python
# Set the URL of the site you're looking to scrape PDFs from (presuming the PDFs are linked directly to on this page)
url = "https://irp.fas.org/doddir/dod/index.html"

# Set a folder to save the downloaded PDFs into
# If there is no such folder, the script will create one automatically
folder_location = 'DoD_Docs_2'
if not os.path.exists(folder_location):os.mkdir(folder_location)
```

```python
# Run this to download PDFs
# Only run it once for the same website/folder (unless you want duplicate copies)

response = requests.get(url)
soup= BeautifulSoup(response.text, "html.parser")     
for link in soup.select("a[href$='.pdf']"):
    #Name the pdf files using the last portion of each link which are unique in this case
    filename = os.path.join(folder_location,link['href'].split('/')[-1])
    with open(filename, 'wb') as f:
        try:
            f.write(requests.get(urljoin(url,link['href'])).content)
            print("{} download completed".format(filename))
        except:
            print("{} did not work due to timeout".format(filename))
```
